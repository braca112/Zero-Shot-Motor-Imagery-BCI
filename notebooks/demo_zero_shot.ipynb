# Demo: Zero-Shot Motor Imagery BCI
This notebook demonstrates a simple workflow for zero-shot EEG motor imagery classification.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load sample data
df = pd.read_csv('../data/sample_eeg.csv')
df.head()
```

## Visualize EEG signals

```python
plt.figure(figsize=(12,6))
for i in range(min(5, df.shape[1])):
    plt.plot(df.iloc[:,i], label=f'Channel {i+1}')
plt.title('Sample EEG Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

## Simulate Zero-Shot Prediction
Since this is a demo, we generate random predictions to simulate a model.

```python
# Simulate 2 classes (left hand / right hand)
predictions = np.random.choice(['Left Hand', 'Right Hand'], size=df.shape[0])
pd.Series(predictions).value_counts()
```
